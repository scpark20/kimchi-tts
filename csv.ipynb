```python
import csv
import numpy as np
```

```python

lines = []
for tag in ['GT', 'S4G', 'S5G', 'W4G', 'W5G']:
    for i in range(100):
        url = 'https://storage.googleapis.com/hvae_tts/exp1/inference_' + str(i) + '_' + tag + '.wav'
        lines.append(url)
        
np.random.shuffle(lines)
print(lines)
```

```python
with open('audio_url.csv', 'w', newline='') as csvfile:
    fieldnames = ['audio_url']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for l in lines:
        writer.writerow({'audio_url': l})
```

```python

lines = []
for tag in ['GT', 'S5G1.0', 'S5G0.666', 'S5G0.333', 
            'W4G1.0', 'W4G0.666', 'W4G0.333']:
    for i in range(100):
        url = 'https://storage.googleapis.com/hvae_tts/exp2/inference_' + str(i) + '_' + tag + '.wav'
        lines.append(url)
        
np.random.shuffle(lines)
print(len(lines))
for line in lines:
    print(line)
```
